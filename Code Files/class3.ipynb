```python
import cv2, mediapipe as mp,numpy as np, matplotlib.pyplot as plt
img = cv2.imread('butterfly.jpg',-1)
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.imshow(img)
plt.title('Loaded Image with OpenCV')
plt.axis('on')
plt.show()
```

```python
height, width, channel =(img.shape)
print(img.shape)
print(img.size)
print(img.dtype)
print(height * width * channel)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title('Gray Scale Image')
plt.axis('off')
plt.show()
```

```python
# Resize
resized = cv2.resize(img, (300, 200))
plt.imshow(resized)
#plt.imshow(img)
plt.title('Loaded Image With OpenCV')
plt.axis('off')
print(resized.shape)
plt.show()
```

```python
# Rotate
rotated = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.imshow(rotated)
#plt.imshow(img)
plt.title('Loaded With OpenCV')
plt.axis('off')
plt.show()
```

```python
# flipped : 0 -> vertical or 1 -> horizontal
flipped = cv2.flip(img, 0)
plt.imshow(flipped)
#plt.imshow(flipped)
plt.title('Loaded With OpenCV')
plt.axis('off')
plt.show()
```

```python
blur = cv2.GaussianBlur(img, (7, 7), 100)
plt.imshow(blur)
#plt.imshow(blur)
plt.title('Loaded With OpenCV')
plt.axis('off')
plt.show()
```

```python
# edges
edges = cv2.Canny(img, 100, 200)
plt.imshow(edges)
#plt.imshow(edges)
plt.title('Loaded With OpenCV')
plt.axis('off')
plt.show()
```

```python
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
plt.imshow(thresh)
#plt.imshow(thresh)
print(ret)
plt.title('Loaded With OpenCV')
plt.axis('off')
plt.show()
 
```

```python
canvas = img.copy()
cv2.rectangle(canvas, (50, 50), (200, 200), (255, 0, 0), 3)
cv2.circle(canvas ,(300, 150), 50, (0, 255, 0), -1)
cv2.putText(canvas,'OpenCv Demo', (100, 300), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255),2)
plt.imshow(canvas)
#plt.imshow(canvas)
plt.title('Loaded With OpenCV')
plt.axis('off')
plt.show()
```

```python
import cv2, mediapipe as mp,numpy as np, matplotlib.pyplot as plt
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Cannot")
else:
    ret, frame = cap.read()
    if ret:
        plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        plt.title("Webcam")
        plt.axis("off")
    cap.release()
```
